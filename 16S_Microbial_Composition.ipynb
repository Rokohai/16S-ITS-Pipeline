```python
# prompt: gzip all the files in Kombucha_Data_Analysis

import os
import gzip

# Define the directory containing the files
directory_to_gzip = "/content/Kombucha_Data_Analysis/Stir16SUnzipped"

# Check if the directory exists
if not os.path.isdir(directory_to_gzip):
    print(f"Directory '{directory_to_gzip}' not found.")
else:
    # Iterate through all files in the directory
    for filename in os.listdir(directory_to_gzip):
        filepath = os.path.join(directory_to_gzip, filename)

        # Check if the current item is a file and not already gzipped
        if os.path.isfile(filepath) and not filepath.endswith('.gz'):
            print(f"Gzipping: {filename}")
            output_filepath = filepath + '.gz'

            try:
                # Open the input file for reading in binary mode
                with open(filepath, 'rb') as f_in:
                    # Open the output file for writing in gzip binary mode
                    with gzip.open(output_filepath, 'wb') as f_out:
                        # Copy data from input to output
                        f_out.writelines(f_in)
                print(f"Gzipped successfully to: {output_filepath}")

                # Optionally, remove the original file after gzipping
                # os.remove(filepath)
                # print(f"Removed original file: {filepath}")

            except Exception as e:
                print(f"Error gzipping {filename}: {e}")
        elif os.path.isfile(filepath) and filepath.endswith('.gz'):
            print(f"Skipping {filename} (already gzipped)")
        elif os.path.isdir(filepath):
            print(f"Skipping directory: {filename}")
        else:
            print(f"Skipping unknown item: {filename}")

    print("Gzipping process finished.")
```
